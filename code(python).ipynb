```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
df.isnull().sum()
df['rate'] = df['rate'].replace('NEW', np.nan)
```

```python
df
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '').astype(float)
```

```python
df
```

```python
df['name'].value_counts()[:10].plot(kind='bar')
```

```python
df['rest_type'].value_counts()[:10].plot(kind='bar')
```

```python
cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(y=cuisines.index, x=cuisines.values)
```

```python
sns.histplot(df['rate'].dropna(), bins=20, kde=True)
```

```python
sns.scatterplot(x='rate', y='approx_cost(for two people)', data=df)
```

```python
location_counts = df['location'].value_counts().head(10)
location_counts.plot(kind='bar')
```

```python
sns.boxplot(data=df, x='online_order', y='rate')
plt.title("Online Order Availability vs Rating")
```

```python
sns.scatterplot(x='votes', y='rate', data=df)
plt.title("Votes vs Rating")
plt.xscale('log')  # if vote values vary widely
```

```python
location_cuisine = pd.crosstab(df['location'], df['cuisines'].str.split(',').str[0])
top_locations = df['location'].value_counts().head(10).index
heatmap_data = location_cuisine.loc[top_locations].iloc[:, :10]  # top 10 cuisines

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt='d')
plt.title("Top Cuisines in Popular Locations")
```

```python
from wordcloud import WordCloud

text = ' '.join(df['cuisines'].dropna().astype(str).tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Cuisines WordCloud")
```

```python
df['book_table'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Dine-In Availability")
```

```python
sns.pairplot(df[['rate', 'votes', 'approx_cost(for two people)']].dropna())
```
